**PROBLEM STATEMENT 1 **- write a SQL query that will return all the opposites of the values in the Value column. If a number is negative, make it positive and vice versa. For this problem statement, your database table is structured like this:

image.png

AND Here’s what your query should return: 56, -76, 84, -96, 47

```python

#CREATING DATABASE 

import sqlite3

conn = sqlite3.connect('Test.db')
print("Opened database successfully");
```

```python

#CREATING TABLE 

conn.execute('''CREATE TABLE IF NOT EXISTS Book(id INTEGER, Value INTEGER);''')
conn.commit()
print("successfully")
```

```python

# INSERTING VALUES - TABLE SCRIPTS

conn.execute("INSERT  INTO Book (id,Value) VALUES (1,-56);")
conn.execute("INSERT INTO Book (id,Value) VALUES (2,76);")
conn.execute("INSERT INTO Book (id,Value) VALUES (3,-84);")
conn.execute("INSERT INTO Book (id,Value) VALUES (4,96);")
conn.execute("INSERT  INTO Book (id,Value) VALUES (5,-47);")
conn.commit()
```

```python
a=conn.execute("Select * from book;")
for r in a:
  print(r)
```

```python
type(a)
```

```python
c=conn.execute('''select id ,value,
case when value>0 then -1*value
when value<0 then abs(value)
end as opp_value
from book;''')
```

```python
for r in c:
  print(r)
```

**PROBLEM STATEMENT 4 **- Even or odd Write a query to return whether or not the number in the Value column is even or odd rather than returning the number itself.Here’s the database table you’ll use in this challenge:

image.png

Your query should return the following values: even, odd, odd, even, odd

```python
conn=sqlite3.connect("Test.db")
```

```python
conn = sqlite3.connect('Test.db')
conn.execute('''CREATE TABLE IF NOT EXISTS Even_Odd(id INTEGER, Value INTEGER);''')
conn.commit()
print("Table created successfully");
```

```python
conn.commit()
```

```python

conn.execute("INSERT INTO Even_odd(id,Value) VALUES(1,4);")
conn.execute("INSERT INTO Even_odd(id,Value) VALUES(2,11);")
conn.execute("INSERT INTO Even_odd(id,Value) VALUES(3,57);")
conn.execute("INSERT INTO Even_odd(id,Value) VALUES(4,24);")
conn.execute("INSERT INTO Even_odd(id,Value) VALUES(5,47);")

conn.commit()
     
```

```python
conn.commit()
```

```python
a = conn.execute(''' 
SELECT DISTINCT id, Value, 
CASE WHEN Value%2 ==0 THEN 'Even' 
else 'Odd' 
END AS category
FROM Even_odd''')

for r in a:
  print(r)
```
